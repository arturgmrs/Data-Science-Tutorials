```
df = pd.read_csv ("/content/DOB_Violations.csv")
df.isnull().sum()
```

```
df["LOT"] = df["LOT"].fillna('*')
df.isnull().sum()
```

```
df["BIN"] = df["BIN"].fillna(df["BIN"].mean())#alterando pelo valor mÃ©dio
df.isnull().sum()
df
```

```
df["DISPOSITION_DATE"] = df["DISPOSITION_DATE"].fillna(df["DISPOSITION_DATE"].interpolate()) #adicionar valores interpolados
df["LOT"] = df["LOT"].fillna('*')
df.isnull().sum()
df
```

```
df = df.dropna() #atenÃ§Ã£o, Ã© para excluir todos os dados
df
```
